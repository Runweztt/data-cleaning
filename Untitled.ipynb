```python
def fill_missing_with_mode(df):
    # Loop through each column in the dataset
    for column in df.columns:
        # Check if the column has categorical data
        if df[column].dtype == 'object':
            mode_value = df[column].mode()[0]
            df[column].fillna(mode_value, inplace=True)
fill_missing_with_mode(df)
```

```python
def fill_missing_numerical(df, strategy='mean'):
    # Loop through each column in the dataset
    for column in df.columns:
        # Check if the column has numerical data
        if pd.api.types.is_numeric_dtype(df[column]):
            if strategy == 'mean':
                fill_value = df[column].mean()
            elif strategy == 'median':
                fill_value = df[column].median()
            else:
                raise ValueError("Invalid strategy. Choose either 'mean' or 'median'.")
            df[column].fillna(fill_value, inplace=True)
fill_missing_numerical(df, strategy='mean')
```
